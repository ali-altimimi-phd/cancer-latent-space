# 01 - Load and inspect VAE inputs

This notebook:

1. Loads the VAE input matrix and metadata exported by the R pipeline
2. Verifies paths, shapes, and sample alignment
3. Confirms the metadata contract used by downstream notebooks


```python
%run ./00_latent_space_config.ipynb

pd.set_option("display.max_columns", 200)
pd.set_option("display.max_rows", 100)
pd.set_option("display.width", 120)
```

```python
expression, metadata = load_ml_inputs(require_full_contract=True)

print("expression shape:", expression.shape)
print("metadata shape  :", metadata.shape)

display(expression.head())
display(metadata.head())
```

```python
missing = [c for c in REQUIRED_METADATA_COLUMNS if c not in metadata.columns]
assert not missing, f"Missing required metadata columns: {missing}"

print("Metadata contract satisfied.")
print(metadata.columns.tolist())
```

```python
features = load_feature_table()
print("Feature table shape:", features.shape)
display(features.head())
```

```python
# Check whether matrix row IDs and metadata row IDs align
same_order = expression.index.equals(metadata.index)
same_set = set(expression.index) == set(metadata.index)

print("Same sample IDs in same order?", same_order)
print("Same sample ID sets?         ", same_set)

if not same_set:
    print("Missing in expression:", sorted(set(metadata.index) - set(expression.index))[:10])
    print("Missing in metadata  :", sorted(set(expression.index) - set(metadata.index))[:10])

assert same_order, "Expression and metadata indices are not aligned."
print("Alignment check passed.")
```

```python
# Confirm explicit sample_id column agrees with row index
sample_id_matches_index = metadata["sample_id"].astype(str).equals(metadata.index.to_series().astype(str))

print("sample_id column matches metadata index?", sample_id_matches_index)
assert sample_id_matches_index, "sample_id column does not match metadata index."
```

```python
print("Disease labels:")
print(sorted(metadata["disease_clean"].unique()))
print()

print("Tissue labels:")
print(sorted(metadata["tissue_clean"].unique()))
print()

print("Condition counts:")
display(metadata["condition"].value_counts(dropna=False))

print("Platform counts:")
display(metadata["platform_id"].value_counts(dropna=False))
```

```python
print("Missing values per metadata column:")
display(metadata.isna().sum().sort_values(ascending=False))

print("Duplicate metadata indices:", metadata.index.duplicated().sum())
print("Duplicate expression indices:", expression.index.duplicated().sum())
```
